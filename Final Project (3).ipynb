```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
train["Arrival Delay in Minutes"].fillna(0, inplace=True)
test["Arrival Delay in Minutes"].fillna(0, inplace=True)
```

```python
train = train.replace({'Loyal Customer':1, 'disloyal Customer':0, 'satisfied': 1, 'neutral or dissatisfied': 0})
train.head()
```

```python
eco = []
eco_plus = []
business = []
for x in train['Class']:
    if x == 'Eco':
        eco.append(1)
        eco_plus.append(0)
        business.append(0)
    elif x == "Business":
        eco.append(0)
        eco_plus.append(0)
        business.append(1)
    else:
        eco.append(0)
        eco_plus.append(1)
        business.append(0)
        
train['eco'] = eco
train['eco_plus'] = eco_plus
train['business'] = business
train.head()
```

```python
business_travel = []
personal_travel = []
for x in train['Type of Travel']:
    if x == 'Personal Travel':
        personal_travel.append(1)
        business_travel.append(0)
    else:
        personal_travel.append(0)
        business_travel.append(1)
train['business_travel'] = business_travel
train['personal_travel'] = personal_travel
train.head()
```

```python
test = test.replace({'Loyal Customer':1, 'disloyal Customer':0, 'satisfied': 1, 'neutral or dissatisfied': 0})
test.head()
```

```python
eco = []
eco_plus = []
business = []
for x in test['Class']:
    if x == 'Eco':
        eco.append(1)
        eco_plus.append(0)
        business.append(0)
    elif x == "Business":
        eco.append(0)
        eco_plus.append(0)
        business.append(1)
    else:
        eco.append(0)
        eco_plus.append(1)
        business.append(0)
        
test['eco'] = eco
test['eco_plus'] = eco_plus
test['business'] = business
test.head()
```

```python
business_travel = []
personal_travel = []
for x in test['Type of Travel']:
    if x == 'Personal Travel':
        personal_travel.append(1)
        business_travel.append(0)
    else:
        personal_travel.append(0)
        business_travel.append(1)
test['business_travel'] = business_travel
test['personal_travel'] = personal_travel
test.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.metrics import auc
```

```python
X_train = train[['Customer Type', 'Inflight wifi service', 'Departure/Arrival time convenient', 
                 'Ease of Online booking', 'Gate location', 'Food and drink', 'Online boarding',
                'Seat comfort', 'Inflight entertainment', 'On-board service', 'Leg room service',
                'Baggage handling', 'Checkin service', 'Inflight service', 'Cleanliness', 
                 'Departure Delay in Minutes', 'Arrival Delay in Minutes', 'eco', 'eco_plus', 'business',
                'business_travel', 'personal_travel']]
X_train.info()
```

```python
y_train = train[['satisfaction']]
y_train.info()
```

```python
X_test = test[['Customer Type', 'Inflight wifi service', 'Departure/Arrival time convenient', 
                 'Ease of Online booking', 'Gate location', 'Food and drink', 'Online boarding',
                'Seat comfort', 'Inflight entertainment', 'On-board service', 'Leg room service',
                'Baggage handling', 'Checkin service', 'Inflight service', 'Cleanliness', 
                 'Departure Delay in Minutes', 'Arrival Delay in Minutes', 'eco', 'eco_plus', 'business',
                'business_travel', 'personal_travel']]
X_test.info()
```

```python
y_test = test[['satisfaction']]
y_test.info()
```

```python
log = LogisticRegression()
log.fit(X_train, y_train)
accuracy = log.score(X_train, y_train) * 100
accuracy = "%.2f" % accuracy
print(f"The training accuracy of logistic regression: {accuracy}%")
```

```python
accuracy = log.score(X_test, y_test) * 100
accuracy = "%.2f" % accuracy
print(f"The testing accuracy of logistic regression: {accuracy}%")
```

```python
from sklearn.neighbors import KNeighborsClassifier

def KNNModel(X_train, y_train, X_test, y_test, n):
    neigh = KNeighborsClassifier(n_neighbors=n).fit(X_train, y_train)
    train_accuracy = neigh.score(X_train, y_train)*100
    train_accuracy = "%.2f" % (train_accuracy)
    print(f"KNN-{n} train accuracy: " + train_accuracy + "%")
    test_accuracy = neigh.score(X_test, y_test)*100
    test_accuracy = "%.2f" % (test_accuracy)
    print(f"KNN-{n} test accuracy: " + test_accuracy + "%")
    
```

```python
y_train_numpy = y_train['satisfaction'].to_numpy()
y_test_numpy = y_test['satisfaction'].to_numpy()
```

```python
KNNModel(X_train, y_train_numpy, X_test, y_test_numpy, 3)
```

```python
KNNModel(X_train, y_train_numpy, X_test, y_test_numpy, 5)
```

```python
KNNModel(X_train, y_train_numpy, X_test, y_test_numpy, 10)
```

```python
KNNModel(X_train, y_train_numpy, X_test, y_test_numpy, 15)
```

```python
KNNModel(X_train, y_train_numpy, X_test, y_test_numpy, 20)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Input
from keras.losses import BinaryCrossentropy
from tensorflow.keras.metrics import Precision
from tensorflow.keras.metrics import Recall
```

```python
model = Sequential()
model.add(Dense(32, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss=BinaryCrossentropy(), optimizer='adam', metrics='accuracy')
model.build(input_shape=(None, 22))
model.summary()
```

```python
model.fit(X_train, y_train, batch_size=128, epochs=5, validation_split=0.2)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score,accuracy_score
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
def process_label(y_pred):
    ans = []
    for y in y_pred:
        if y > 0.5:
            y = 1
        else:
            y = 0
        ans.append(y)
    return ans
```

```python
y_pred = process_label(y_pred)
```

```python
precision = precision_score(y_test, y_pred, average='binary')*100
precision = "%.2f" % (precision)
print(f"Precision: " + precision + "%")
```

```python
recall = recall_score(y_test, y_pred, average='binary')*100
recall = "%.2f" % (recall)
print(f"Recall: " + recall + "%")
```

```python
f1_score = f1_score(y_test, y_pred, average='binary')*100
f1_score = "%.2f" % (f1_score)
print(f"F1 score: " + f1_score + "%")
```

```python
accuracy = accuracy_score(y_test, y_pred)*100
accuracy = "%.2f" % (accuracy)
print(f"Accuracy: " + accuracy + "%")
```
